```python
def isPalindrome(s: str) -> bool:
    s1 = s.lower()
    stringnew = []
    for i in s1:
        if i.isalnum():
            stringnew += i

    def checkPoli(wor):
            return wor == wor[::-1] 

    return checkPoli(stringnew)
```
