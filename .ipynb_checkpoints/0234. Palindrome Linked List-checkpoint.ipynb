234. Palindrome Linked List

Given the head of a singly linked list, return true if it is a palindrome or false otherwise.

Example 1:


```python
def isPalindrome(head) -> bool:
    
    return head == head[::-1]
```

```python
head = [1,2,3,3,2,1]
isPalindrome(head)
```

```python
head = [1,2,3,4,5,6]
isPalindrome(head)
```
